```python
# function with no argument and return type
def cal():
    l = 10
    b = 5
    h = 3
    a = l*b
    v = a*h
    return a,v


x = cal()
area,volume = x
print(x)
print("The area is = ",area)
print("The volume is = ",volume)
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b

a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("+ - * /  ")
if o == "+":
    add(a,b)
    
elif o == "-":
    sub(a,b)
    
elif o == "*":
    print(a*b)

elif o =="/":
    print(a/b)
    
else:
    print("invalid operator")
```

```python
def area(l,b):
    return l*b

def volume(l,b,h):
    v = area(l,b)*h
    return v

l = 10
b = 4
h = 5

print(area(l,b))
print(volume(l,b,h))
```
