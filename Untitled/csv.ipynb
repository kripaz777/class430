```python
b = open('data.csv','a')
n = int(input("Enter n = "))
b.write("Sn,Name,Age,Address,Phone\n")
for i in range(n):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    address = input("Entre address = ")
    phone = input("Enter phone = ")
    data = f"{i},{name},{age},{address},{phone}\n"
    b.write(data)
    
b.close()
```
